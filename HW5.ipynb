##1.Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x


```python
a = int  (input())
def amstrong(a):
    tong = 0
    degit = str(a)
    n = len(degit)
    for  i in degit:
        tong += int(i) ** n
    if tong == a:
            return (f'{a} la so amstrong')
    else:
        return (f'{a} khong la so amstrong')
print (amstrong(a))
```

##VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
n = int(input())
def sokidieu(n):
    a = str(n)
    b = len(a)
    tong = 0
    for i in a:
        tong += int(i) ** b
    if tong == n:
        return (f'{n} la so ki dieu')
    else:
        return (f'{n} khong la so ki dieu')
print (sokidieu(n))
```

##Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
n = input()
def detect_number_type (n):
    if '.' in str(n):
        return ('float')
    elif int(n) >= 0:
        return ("int")
    else:
        return "neither"
print (detect_number_type(n))
```

##Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
n =  int (input())
def is_one_product_of_others(n):
    for i in str(n):
        for j in str(n):
            for k in str(n):
                if int(i) * int(j) == int(k):
                    return True
    return False
print (is_one_product_of_others(n))
```

##Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    if (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1):
        return True
    else:
        return False
x1, y1, x2, y2, x3, y3 = map(int, input().split())
print (are_collinear(x1, y1, x2, y2, x3, y3))
```

##Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
n = int(input())
def sochusokhac0(n):
    count = 0
    for  i in str(abs(n)) :
            if i != '0':
                count += 1
    return count
print(sochusokhac0(n))
```

##Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3)

##Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]

```python
a,b, c = map(int, input().split())
def maxmin(a, b, c):
    max_value = max(a, b, c)
    min_value = min(a, b, c)
    return max_value, min_value
print(maxmin(a, b, c))
```

##Viết hàm tính hệ số nhị thức C(n, k).


```python
n = int(input())
k = int(input())
def C(n, k):
    if k > n or k < 0:
            return 0
    if k == 0 or k == n:
            return 1
    res = 1
    for i in range (1, 1+k):
        res = res * (n - i +1)//i
    return res
print (C(n, k))
 
```
