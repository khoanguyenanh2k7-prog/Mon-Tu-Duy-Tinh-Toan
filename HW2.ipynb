##Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.


```python
a,b,c = map(float, input ("Nhap 3 canh cua tam giac: ").split())
p = (a+b+c)/2
if a+b<=c and b+c<=a and a+c<=b:
    print ("Day khong phai la 3 canh cua tam giac")
elif a==b and b==c:
    print ("Day la tam giac deu.", "Dien tich cua tam giac deu la: ", (p*(p-a)*(p-b)*(p-c))**0.5, ".Chu vi cua tam giac deu la: ", a+b+c)
elif a==b or a==c or b==c:
    print ("Day la tam giac can.", "Dien tich cua tam giac can la: ", (p*(p-a)*(p-b)*(p-c))**0.5, ".Chu vi cua tam giac can la: ", a+b+c)
elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
    print ("Day la tam giac vuong.", "Dien tich cua tam giac vuong la: ", (p*(p-a)*(p-b)*(p-c))**0.5, ".Chu vi cua tam giac vuong la: ", a+b+c)
else:
    print ("Day la tam giac thuong.", "Dien tich cua tam giac thuong la: ", (p*(p-a)*(p-b)*(p-c))**0.5, ".Chu vi cua tam giac thuong la: ", a+b+c)
```

##Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0)

```python
a,b = map(float, input("Nhap 2 so: ").split())
if a==0 and b==0:
    print("Phuong trinh co vo so nghiem")
else:
    x = -b/a
    print ("Nghiem cua phuong trinh bac nhat la x=", x)
```

##Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a,b,c = map(float, input ("Nhap 3 so: ").split())
delta = b**2 - 4*a*c
if delta < 0:
    print ("Phuong trinh vo nghiem")
elif delta == 0:
    x = -b/(2*a)
    print ("Phuong trinh co nghiem kep x=", x)
else:
    x1 = (-b + delta**0.5)/(2*a)
    x2 = (-b - delta**0.5)/(2*a)
    print ("Phuong trinh co 2 nghiem phan biet x1=", x1, "va x2=", x2)
```
