##Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải


```python
n, m = map(int, input().split())

a = []

for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)

for row in a:
    for x in row:
        print(f"{x:>4}", end="")
    print()
```

##Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k) 


```python
n, m, k = map(int, input().split())
a = []
k = k -1 
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)
print ("Tong cac phan tu tren cot k: ", end="")
tongcotk = 0
for i in range(n):
    tongcotk += a[i][k]
print (tongcotk)
print ()
print ("Tong cac phan tu tren hang k: ", end="")
tonghangk = 0
for j in range(m):
    tonghangk += a[k][j]
print (tonghangk)
print()
```

##Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1


```python
nums = list(map(int, input().split()))
k = int(input())

def result(nums, k):
    for i in range(len(nums)):
        if nums[i] == k:
            return i+1

    return -1

print(result(nums, k))
```

##Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
s = input().split()

d = {}

for i in range(len(s)):
    if s[i] not in d:
        d[s[i]] = (i,)
    else:
        d[s[i]] += (i,)

print(d)
```

##Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
s = list(map(int, input().split()))
def max(s):
    max_value = s[0]
    for i in range(1, len(s)):
        if s[i] > max_value:
            max_value = s[i]
    return i
print(max(s))
```

##VIết hàm xóa đi các số <x có trong mảng

```python
s = list(map(int, input().split()))
x = int(input())
s1 = []
def xoasonhohonx(s, x):
    for i in s:
        if i >= x:
            s1.append(i)
    return s1

print(xoasonhohonx(s, x))
```

##Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
s = list(map(int, input().split()))
s1 = []
s2 = []
def chuyen0lendauvacuoi(s):
    for i in s:
        if i == 0:
            s1.append(i)
        else:
            s2.append(i)
    return (s1+s2, s2+s1)

print(chuyen0lendauvacuoi(s))
```

##Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
nums = list(map(int, input().split()))
def tong(nums):
    tong = 0
    for i in range(1, len(nums)-1):
        if nums[i] > nums[i-1] and nums[i] > nums[i+1]:
            tong += nums[i]
    return tong
print(tong(nums))
```

##Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1


```python
def vi_tri_nho_hon_x(arr, x):
    vi_tri = []

    for i in range(len(arr)):
        if arr[i] < x:
            vi_tri.append(i)

    if len(vi_tri) == 0:
        print(-1)
    else:
        print(*vi_tri)


arr = list(map(int, input().split()))
x = int(input())
vi_tri_nho_hon_x(arr, x)
```

##Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
nums = list(map(int, input().split()))
def check(nums):
    for i in range (len(nums)):
        if nums[i] > nums[i+1]:
            return ("day tang")
        elif nums[i] < nums[i+1]:
            return ("day giam")
        else:
            return("day vo danh")
print (check(nums))
```
