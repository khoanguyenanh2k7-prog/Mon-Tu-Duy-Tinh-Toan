Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
nums = list(map(int, input().split()))
result  = []
for i in nums:
    if i not in result:
        result.append(i)
print (result)
```

##W6A2
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự

```python
nums = list (map(int, input().split()))
result = []
s = 0
for i in nums:
    s += i
    result.append(s)
print (result)
```

##Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
lặp lại, thêm vào list của key đó


```python
a = input("Nhập các cặp key:value: ").split()
result = {}

for i in a:
    if ':' in i: 
        key, value = i.split(':')
        
        if key not in result:
            result[key] = [] 
            
        result[key].append(value) 

print(result)
```

##Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)

```python
nums = list(map(int, input().split()))
dict = {}
for i in nums:
    if i > 0:
        dict[i] = 'positive'
    elif i < 0:
        dict[i] = 'negative'
    else:
        dict[i] = 'zero'
print (dict)
```

##Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện
của từng chuỗi và lưu vào dictionary

```python
chuoi = list(input().split())
counts = {}
for i in chuoi:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1
print (counts)
```
