##9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
n = int(input("Nhap mot so nguyen tu 0 den 9: "))
match n:
    case 0:
        print ("Khong")
    case 1:
        print ("Mot")
    case 2:
        print ("Hai")
    case 3:
        print ("Ba")
    case 4:
        print ("Bon")
    case 5:
        print ("Nam")
    case 6:
        print ("Sau")
    case 7:
        print ("Bay")
    case 8:
        print ("Tam")
    case 9:
        print ("Chin")
```

##Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a,b,c= map(float, input().split())
if a+b>c and a+c>b and b+c>a:
    print("Dien tich cua tam giac la: ", ((a+b+c)/2*(a+b+c-2*a)*(a+b+c-2*b)*(a+b+c-2*c))**0.5)
else:
    print("Khong phai la 3 canh cua tam giac")
```
