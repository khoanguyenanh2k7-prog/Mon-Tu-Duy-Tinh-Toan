```python
a,b,c= map(float, input().split())
if a+b>c and a+c>b and b+c>a:
    print("Dien tich cua tam giac la: ", ((a+b+c)/2*(a+b+c-2*a)*(a+b+c-2*b)*(a+b+c-2*c))**0.5)
else:
    print("Khong phai la 3 canh cua tam giac")
```

###Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet. 

```python
c = input()
if c.isupper():
    print (f"{c} la ki tu alphabet")
else:
    print (f"{c} khong la ki tu alphabet")
```
